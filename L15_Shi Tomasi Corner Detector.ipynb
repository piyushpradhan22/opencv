```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
kernel = np.ones((3,3), np.uint8)
img = cv2.imread('data/chess.jpg',1)
imgray = np.float32(cv2.imread('data/chess.jpg',0))
corners = cv2.goodFeaturesToTrack(imgray, 100,0.1,10) #img, max corners, quality, min distance
corners = np.int0(corners)
```

```python
for corner in corners:
    x,y = corner.ravel()
    cv2.circle(img, (x,y), 10, (255,255,0),2)
```

```python
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```
