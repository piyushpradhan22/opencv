```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
```

```python
kernel = np.ones((3,3), np.uint8)
img = cv2.imread('data/chess.jpg',1)
imgray = np.float32(cv2.imread('data/chess.jpg',0))
corners = cv2.cornerHarris(imgray, 5, 5, 0.04)
dst = cv2.dilate(corners,kernel,iterations=2)
```

```python
plt.imshow(dst, cmap='gray')
```

```python
img[dst > .1 * np.max(dst)] = [255,0,0]
```

```python
plt.imshow(cv2.cvtColor(img,cv2.COLOR_BGR2RGB))
```
