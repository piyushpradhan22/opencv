```python
import cv2
import numpy as np
import time
import matplotlib.pyplot as plt
```

```python
cap = cv2.VideoCapture('exercise.mp4')
n=0
while(cap.isOpened()):
    n+=1
    ret, frame = cap.read()
    if ret == False or (cv2.waitKey(1) & 0xFF == ord('q')) or n==60:
        img = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        break
    cv2.imshow('Tracking',cv2.resize(cv2.flip(frame,1),(480,640)))
cv2.destroyAllWindows()
cap.release()
```

```python
plt.imshow(img)
plt.show()
```
